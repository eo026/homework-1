```python
import pandas as pd

import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://fivethirtyeight.com/features/can-focusing-on-climate-change-help-win-elections/'
```

```python
html=requests.get(url)
```

```python
html.status_code
```

```python
assert html.status_code == 200
assert html.url == "https://fivethirtyeight.com/features/can-focusing-on-climate-change-help-win-elections/"
```

```python
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
quotes_all = soup.find_all('div', {'class': 'entry-content'})
```

```python
quotes_all
```

```python
quotes_text=[]
for i in quotes_all:
    quotes_text.append((i.find("p",{"data-paragraph":"main"})).text)
```

```python
quotes_text
```

```python
df_quotes = pd.DataFrame({"content" : quotes_text})
df_quotes
```
